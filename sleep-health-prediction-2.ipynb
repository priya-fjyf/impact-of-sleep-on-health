## Import Necessery Algorithms 

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import category_encoders 
from category_encoders import TargetEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

```

### Remove first Column

```python
df = df.drop('Person ID', axis=1)
```

```python
df.info()
```

## Separate the Independent columns and dependent columns 

```python
X = df.iloc[:,0:11]
y = df.iloc[:,-1]
```

### Encode the Target Variable

```python
label = LabelEncoder()
encoded_y = label.fit_transform(y)
```

## Separate the Dataset Into Training and Testing Dataset

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.1,random_state =42)
```

```python
X_train
```

```python
y_train
```

```python
y_test
```

```python
df.info()
```

## Encode the Variable Using Column Transformer


```python
numeric_cols = ['Age','Sleep Duration','Quality of Sleep','Physical Activity Level','Stress Level','Heart Rate',
               'Daily Steps']
categoric_cols = ['Gender','Occupation']
ordinal_cols = ['BMI Category','Blood Pressure']

```

```python
categories = ['Normal','Normal Weight','Overweight','Obese']
bp_categories = [
    '115/75',
    '120/80',
    '125/80',
    '129/84',
    '130/85',
    '130/86',
    '131/86',
    '132/87',
    '135/88',
    '135/90',
    '139/91',
    '140/90',
    '140/95',
    '142/92'
]
```

```python
one = OneHotEncoder(handle_unknown='ignore',drop='first')
ordinal =OrdinalEncoder(categories = [categories,bp_categories]) 
scaler = StandardScaler()
```

## Use ColumnTransformer For applying Transformation in Each Column

```python
preprocessor = ColumnTransformer(
    transformers = [
        ('one_hot',one,categoric_cols),
        ('ordinal',ordinal,ordinal_cols),
        ('scaling',scaler,numeric_cols)
    ]
)
```

```python
lr = LogisticRegression()
dt = DecisionTreeClassifier(criterion='gini',splitter='best',max_depth=None)
rf = RandomForestClassifier(n_estimators=100,criterion='gini',max_depth=None)
svc = SVC(C=1.0, kernel='linear', degree=3, gamma='scale', shrinking=True, probability=False)
```

## Use Pipeline for Fit the model 

```python
pipeline = Pipeline([
    ('preprocessor',preprocessor),
    ('lr',lr)
])
```

```python
pipeline.fit(X_train,y_train)
```

```python
y_pred =  pipeline.predict(X_test)
```

# Check Accuracy for each Algorithms and Choose the best model

```python
from sklearn.metrics import classification_report,accuracy_score
accuracy_score(y_test,y_pred)
                     
```
